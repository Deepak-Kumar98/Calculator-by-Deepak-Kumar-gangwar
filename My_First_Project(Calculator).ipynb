```python
from  tkinter import*

def click(event):
    global scvalue
                #cget()- how to get text from widget
    text = event.widget.cget("text") # Give the Button which click on that.
    print(text)
    if text == "=":
        if scvalue.get().isdigit():
            value = int(scvalue.get())
        else:
            try:
                value = eval(screen.get())
            except Exception as e:
                #print(e) # To see error on consol
                value="Error"
                screen.update()
            
        scvalue.set(value)
        screen.update()
        pass
    elif text == "C":
        scvalue.set("")
        screen.update()
        pass
    else:
        scvalue.set(scvalue.get() + text)
        screen.update()

root = Tk()

root.geometry("644x900")
root.title("Calculator by Deepak Gangwar")


scvalue = StringVar()
scvalue.set("")
screen = Entry(root, textvar=scvalue, font="lucida 40 bold")
screen.pack(fill=X, ipadx=8, pady=10, padx=10)

f = Frame(root, bg = "grey") # frame=f
b = Button(f, text="9", font ="lucida 35 bold")
b.pack(side = LEFT, padx = 18, pady = 5)
b.bind("<Button-1>", click)
b = Button(f, text="8", font ="lucida 35 bold")
b.pack(side = LEFT, padx = 18, pady = 5)
b.bind("<Button-1>", click)
b = Button(f, text="7",font ="lucida 35 bold")
b.pack(side = LEFT, padx = 18, pady = 5)
b.bind("<Button-1>", click)
f.pack()

f = Frame(root, bg = "grey") # frame=f
b = Button(f, text="6", font ="lucida 35 bold")
b.pack(side = LEFT, padx = 18, pady = 5)
b.bind("<Button-1>", click)
b = Button(f, text="5",font ="lucida 35 bold")
b.pack(side = LEFT, padx = 18, pady = 5)
b.bind("<Button-1>", click)
b = Button(f, text="4", font ="lucida 35 bold")
b.pack(side = LEFT, padx = 18, pady = 5)
b.bind("<Button-1>", click)
f.pack()

f = Frame(root, bg = "grey") # frame=f
b = Button(f, text="3",font ="lucida 35 bold")
b.pack(side = LEFT, padx = 18, pady = 5)
b.bind("<Button-1>", click)
b = Button(f, text="2", font ="lucida 35 bold")
b.pack(side = LEFT, padx = 18, pady = 5)
b.bind("<Button-1>", click)
b = Button(f, text="1", font ="lucida 35 bold")
b.pack(side = LEFT, padx = 18, pady = 5)
b.bind("<Button-1>", click)
f.pack()

f = Frame(root, bg = "grey") # frame=f
b = Button(f, text="0",padx=3, font ="lucida 35 bold")
b.pack(side = LEFT, padx = 18, pady = 5)
b.bind("<Button-1>", click)
b = Button(f, text="-",padx =3, font ="lucida 35 bold")
b.pack(side = LEFT, padx = 18, pady = 5)
b.bind("<Button-1>", click)
b = Button(f, text="*",padx =6, font ="lucida 35 bold")
b.pack(side = LEFT, padx = 18, pady = 5)
b.bind("<Button-1>", click)
f.pack()

f = Frame(root, bg = "grey") # frame=f
b = Button(f, text="/",padx=3, font ="lucida 35 bold")
b.pack(side = LEFT, padx = 18, pady = 5)
b.bind("<Button-1>", click)
b = Button(f, text="+",padx=2,font ="lucida 35 bold")
b.pack(side = LEFT, padx = 18, pady = 5)
b.bind("<Button-1>", click)
b = Button(f, text="=",font ="lucida 35 bold")
b.pack(side = LEFT, padx = 18, pady = 5)
b.bind("<Button-1>", click)
f.pack()

f = Frame(root, bg = "grey") # frame=f
b = Button(f, text="C",padx=1, font ="lucida 35 bold")
b.pack(side = LEFT, padx = 18, pady = 5)
b.bind("<Button-1>", click)
b = Button(f, text="%", font ="lucida 35 bold")
b.pack(side = LEFT, padx = 18, pady = 5)
b.bind("<Button-1>", click)
b = Button(f, text=".", font ="lucida 35 bold")
b.pack(side = LEFT, padx = 18, pady = 5)
b.bind("<Button-1>", click)
f.pack()
root.mainloop()
```
